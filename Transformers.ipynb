```python
q_a = pipeline('question-answering')

context = """
Flipkart is an e-commerce company based in India.
The OnePlus 10R is a smartphone available on Flipkart.
It features a MediaTek Dimensity 8100-Max processor,
a 4500mAh battery with 150W SUPERVOOC Endurance Edition charging,
and a 6.7-inch 120Hz AMOLED display.
The phone has a triple rear camera setup with a 50MP primary sensor.
It runs on OxygenOS based on Android.
Different variants with varying RAM and storage options are available on Flipkart.
"""

print("Context about Flipkart and OnePlus 10R is loaded.")
print("Ask me questions about the OnePlus 10R. Type 'quit' to exit.")

while True:
    question = input("Enter your question: ")
    if question.lower() == 'quit':
        break

    result_txt = q_a({
        'question': question,
        'context': context
    })

    print("Answer:", result_txt['answer'])
    print("-" * 20) # Separator for clarity

print("Exiting the question-answering session.")
```
